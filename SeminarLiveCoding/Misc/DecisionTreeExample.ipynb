```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2)
```

```python
X.shape, y.shape
```

```python
_ = plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], s=100, alpha=0.5, c=y)
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend(["X1", "X2"])
plt.show()
```

```python
# Cum trec de la N dimensiuni la 2 sau 3: Dimensionality reduction, de ex. PCA, UMAP, ...
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,10))
plot_tree(model, ax=ax)
plt.show()
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_acc = accuracy_score(y_true=y_train, y_pred=y_train_pred)
test_acc = accuracy_score(y_true=y_test, y_pred=y_test_pred)

print(f"Train accuracy: {train_acc}.")
print(f"Test accuracy: {test_acc}.")
```

```python
_ = plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], s=25, alpha=1, c=y, label="Real")
plt.scatter(X_train[:, 0], X_train[:, 1], s=150, alpha=0.5, c=y_train, label="Train predictions")
plt.scatter(X_test[:, 0], X_test[:, 1], s=150, alpha=0.5, c=y_test, label="Test predictions")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```
