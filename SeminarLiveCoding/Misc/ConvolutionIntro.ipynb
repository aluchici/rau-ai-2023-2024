```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
image = np.ones((20, 20))
image[5:7, 14:18] = 0
image[7:15, 15:17] = 0
image[15:17, 14:18] = 0 

plt.imshow(image)
plt.show()
```

```python
# blur_3x3 = np.ones((3, 3)) * 1.0/9.0
blur_3x3 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]) # df/dy
blur_3x3 = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]) # df/dx

blur_3x3
```

```python
blur_image = np.zeros(image.shape)

x_indices = [-1, 0, 1]
y_indices = [-1, 0, 1]

for y in range(1, image.shape[0]-1):
    for x in range(1, image.shape[1]-1):
        for i in range(blur_3x3.shape[0]):
            for j in range(blur_3x3.shape[1]):
                yi = y + y_indices[i]
                xi = x + x_indices[j]
                blur_image[y, x] += image[yi, xi] * blur_3x3[i, j]
```

```python
plt.imshow(blur_image)
plt.show()
```
